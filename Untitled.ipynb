```python
pca.fit(scaled_data)
```

```python
pca_data = pca.transform(scaled_data)
```

```python
per_var = np.round(pca.explained_variance_ratio_ * 100,2)
```

```python
labels = ['PC' + str(i+1) + ' ' + df.columns[i] for i in range(len(per_var))]
```

```python
fig, ax = plt.subplots(figsize=(8, 6)) # Adjust the size
ax.bar(labels, per_var)

# Add x-label and y-label
ax.set_xlabel('PC')
ax.set_ylabel('% of variance explained ')
ax.set_xticklabels(labels, rotation=45)
plt.show()
```

#### Great - this tells us that the principal components are PC1 and PC2. We can now proceed to making a scatter plot with these principal components. 

```python
pca_df = pd.DataFrame(data = pca_data, columns = labels)
```
